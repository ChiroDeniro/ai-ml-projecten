```python
from transformers import pipeline
import warnings
warnings.filterwarnings('ignore')

print("Loading 3 different AI models...\n")

# Model 1: Sentiment Analysis
print("1️⃣ Loading Sentiment Analyzer...")
sentiment = pipeline("sentiment-analysis")

# Model 2: Text Summarization
print("2️⃣ Loading Summarizer...")
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Model 3: Question Answering
print("3️⃣ Loading Q&A System...")
qa = pipeline("question-answering")

print("\n✅ All models loaded! Ready to analyze text.\n")
```

```python
# Test text (use something from your econometrics world!)
article = """
The European Central Bank raised interest rates by 50 basis points in its
latest meeting, marking the seventh consecutive increase as policymakers
battle persistent inflation. The decision brings the deposit rate to 3.5%,
the highest level in over a decade. ECB President Christine Lagarde stated
that while inflation has begun to moderate, it remains significantly above
the 2% target. Markets had widely anticipated this move, with economists
predicting further tightening may be necessary in the coming months.
"""

print("=" * 70)
print("MULTI-MODEL TEXT ANALYSIS DEMO")
print("=" * 70)

# Analysis 1: Sentiment
print("\n📊 SENTIMENT ANALYSIS")
print("-" * 70)
result = sentiment(article)[0]
print(f"Overall Tone: {result['label']} ({result['score']:.1%} confidence)")

# Analysis 2: Summarization
print("\n📝 SUMMARIZATION")
print("-" * 70)
summary = summarizer(article, max_length=50, min_length=20, do_sample=False)[0]
print(f"Key Point: {summary['summary_text']}")

# Analysis 3: Question Answering
print("\n❓ QUESTION ANSWERING")
print("-" * 70)

questions = [
    "What did the ECB do?",
    "What is the current deposit rate?",
    "Who is the ECB President?"
]

for q in questions:
    answer = qa(question=q, context=article)
    print(f"Q: {q}")
    print(f"A: {answer['answer']} (confidence: {answer['score']:.1%})\n")
```
