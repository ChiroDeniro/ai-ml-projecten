```python
from transformers import pipeline
import warnings
warnings.filterwarnings('ignore')

# Load model (first time takes ~1 minute to download)
print("Loading sentiment analysis model...")
sentiment_analyzer = pipeline("sentiment-analysis")
print("✅ Model loaded!\n")

# Function to analyze sentiment
def analyze_text(text):
    result = sentiment_analyzer(text)[0]
    label = result['label']
    score = result['score']

    # Pretty print
    emoji = "😊" if label == "POSITIVE" else "😞"
    print(f"{emoji} {label} (confidence: {score:.1%})")
    print(f"Text: '{text}'\n")
    return result

# Test examples
test_texts = [
    "I love this! It's amazing and wonderful!",
    "This is terrible. Worst experience ever.",
    "The weather is okay, I guess.",
    "Absolutely brilliant work! Very impressed!",
    "I'm disappointed with the results."
]

print("=" * 50)
print("SENTIMENT ANALYSIS DEMO")
print("=" * 50 + "\n")

for text in test_texts:
    analyze_text(text)
```

```python
# Interactive mode - try your own text!
print("🎯 Interactive Sentiment Analyzer")
print("=" * 50)

def interactive_analysis():
    while True:
        text = input("\n📝 Enter text (or 'quit' to stop): ")

        if text.lower() in ['quit', 'exit', 'q']:
            print("👋 Goodbye!")
            break

        if text.strip():
            analyze_text(text)
        else:
            print("Please enter some text!")

# Uncomment next line to use interactive mode:
# interactive_analysis()

# For demo, let's analyze some economics-related text!
economics_texts = [
    "The GDP growth exceeded expectations this quarter.",
    "Inflation remains a significant concern for policymakers.",
    "The stock market showed strong performance today.",
    "Unemployment rates have reached concerning levels."
]

print("\n📊 Analyzing Economics Headlines:")
print("=" * 50 + "\n")

for text in economics_texts:
    analyze_text(text)
```
